```python
# Créez un DataFrame pour la population urbaine
urban_data = population[population['Granularity'] == 'Urban']

# Créez un DataFrame pour la population totale
total_data = population[population['Granularity'] == 'Total']

# Créez un dictionnaire pour stocker les ratios année par année
population_ratios = {}

# Parcourez chaque année unique dans le DataFrame
for year in population['Year'].unique():
    # Filtrez les données pour l'année en cours
    urban_year = urban_data[urban_data['Year'] == year]
    total_year = total_data[total_data['Year'] == year]

    # Calculez le rapport pour l'année en cours en pourcentage
    ratio = (urban_year['Population'].sum() / total_year['Population'].sum()) * 100

    # Stockez le rapport dans le dictionnaire
    population_ratios[year] = ratio

# Ajoutez la colonne 'Rapport Population Urbaine/Total' au DataFrame original
population['Rapport Population Urbaine/Total'] = population['Year'].map(population_ratios)

# Affichez le DataFrame avec la nouvelle colonne
population.head(30)
```

```python
# Définissez les poids pour chaque colonne
poids_colonne1 = 0.5
poids_colonne2 = 0.3
poids_colonne3 = 0.3

# Calculez la note en utilisant la moyenne pondérée
total_data_df['Note_Moyenne_Pondérée'] = (total_data_df['Political_Stability'] * poids_colonne1 +
                                        total_data_df['Population using at least basic drinking-water services (%)'] * poids_colonne2 +
                                        total_data_df['Population using safely managed drinking-water services (%)'] * poids_colonne3) * 5

# Affichez le DataFrame avec la colonne de notes basée sur la moyenne pondérée
total_data_df.head()
```

```python
# Créez une nouvelle colonne pour le classement de chaque colonne
total_data_df['Classement_Political_Stability'] = pd.qcut(total_data_df['Political_Stability'], q=5, labels=False) + 1
total_data_df['Classement_Population using at least basic drinking-water services (%)'] = pd.qcut(total_data_df['Population using at least basic drinking-water services (%)'], q=5, labels=False) + 1
total_data_df['Classement_Population using safely managed drinking-water services (%)'] = pd.qcut(total_data_df['Population using safely managed drinking-water services (%)'], q=5, labels=False) + 1

# Calculez la note globale en faisant la somme des classements
total_data_df['Note_Classement'] = total_data_df['Classement_Political_Stability'] + total_data_df['Classement_Population using at least basic drinking-water services (%)'] + total_data_df['Classement_Population using safely managed drinking-water services (%)']

# Affichez le DataFrame avec la colonne de notes basée sur le classement
total_data_df.head()
```

Valeur Normalisée = (Valeur Originale - Min) / (Max - Min)


```python
# Supposons que votre DataFrame pour la mortalité est nommé "Dataframe_mortalité"
weight_mortality = 0.5  # Poids pour le taux de mortalité
weight_water_access = 0.5  # Poids pour l'accès à l'eau potable

# Normalisez les indicateurs de mortalité et d'accès à l'eau potable
Dataframe_mortalité['Normalized Mortality'] = (Dataframe_mortalité['Mortality rate attributed to exposure to unsafe WASH services'] - Dataframe_mortalité['Mortality rate attributed to exposure to unsafe WASH services'].min()) / (Dataframe_mortalité['Mortality rate attributed to exposure to unsafe WASH services'].max() - Dataframe_mortalité['Mortality rate attributed to exposure to unsafe WASH services'].min())
Dataframe_mortalité['Normalized Water Access'] = (Dataframe_mortalité['Population using at least basic drinking-water services (%)'] - Dataframe_mortalité['Population using at least basic drinking-water services (%)'].min()) / (Dataframe_mortalité['Population using at least basic drinking-water services (%)'].max() - Dataframe_mortalité['Population using at least basic drinking-water services (%)'].min())

# Créez la métrique composite "Politique Efficace"
Dataframe_mortalité['Politique Efficace'] = (weight_mortality * Dataframe_mortalité['Normalized Mortality']) + (weight_water_access * Dataframe_mortalité['Normalized Water Access'])
```
