## Population urbaine

```python
population.head()
```

```python
# Regrouper les données par pays, année et Granularity, puis calculer la somme de la population
grouped = population.groupby(['Country', 'Year', 'Granularity'])['Population'].sum().unstack()

# # Calculer la part de la population urbaine sur la population totale
grouped['Urban Population Share (%)'] = round((grouped['Urban'] / grouped['Total']) *100 ,2)

# Réinitialiser l'index
grouped.reset_index(inplace=True)

# Fusionner les données calculées avec le DataFrame d'origine
population = pd.merge(population, grouped[['Country', 'Year', 'Urban Population Share (%)']], on=['Country', 'Year'], how='left')

# Afficher le DataFrame avec la nouvelle colonne
population.tail()
```

## merge avec basic and safety

```python
basic_and_safety.head()
```

```python
merged_data = population.merge(basic_and_safety, on=["Country", "Granularity", "Year"], how='left')
merged_data.head(10)
```

## merge avec political stability

```python
political_stability.head(10)
```

```python
merged_data = merged_data.merge(political_stability, on=["Country", "Year", "Granularity"], how='left')
merged_data.head(10)
```

## Selection de la granularité totale

```python
# Sélectionnez les lignes où 'Granularity' est égal à 'Total'
total_data_df = merged_data[merged_data['Granularity'] == 'Total']
# Affichez le DataFrame avec seulement les lignes 'Granularity' égales à 'Total'
total_data_df.head()
```

## Jointures country avec le reste

```python
region_country.head(10)
```

```python
chine = region_country[region_country['COUNTRY (DISPLAY)']== 'China']
chine
```

```python
total_data_df = total_data_df.merge(region_country, left_on="Country", right_on="COUNTRY (DISPLAY)", how='left')
total_data_df.head()
```

## Création des part de la pop ayant accès et accès safe à l'eau et non accès à l'eau

```python
total_data_df["part de la population ayant accès à l'eau"]= round(total_data_df['Population'] *( total_data_df['Population using at least basic drinking-water services (%)']/100),2)
total_data_df.head()
```

```python
total_data_df["part de la population ayant accès sécurisé à l'eau"]= round(total_data_df['Population'] * (total_data_df['Population using safely managed drinking-water services (%)']/100),2)
total_data_df.head()
```

```python
total_data_df["difference entre accès et accès safe"]= round((total_data_df["part de la population ayant accès à l'eau"] - total_data_df["part de la population ayant accès sécurisé à l'eau"]),2)
total_data_df.head()
```

```python
total_data_df["part de la population n'ayant pas un accès à l'eau"]= round(total_data_df["Population"] -(total_data_df["difference entre accès et accès safe"] + total_data_df["part de la population ayant accès sécurisé à l'eau"]),2)
total_data_df.head()
```

## Normalisation des données


```python
# x_normalized = (x - min(colonne)) / (max(colonne) - min(colonne))
# x_normalized = (x - moyenne(colonne)) / écart-type(colonne)

# Créez un MinMaxScaler pour normaliser les colonnes
scaler = MinMaxScaler()

# Normalisez les colonnes "col1" et "col2"
columns_to_normalize = ['Political_Stability', 'Urban Population Share (%)']
total_data_df[columns_to_normalize] = scaler.fit_transform(total_data_df[columns_to_normalize])

total_data_df.head()
```

# Création du deuxième csv avec un tri sur l'année 2016 afin de faire la jointure avec les données de mortalités sur cette année

## jointure avce le fichier de mortalité

```python
# selection de l'année 2016
Dataframe_mortalité_2016_total = total_data_df[(total_data_df['Year'] == 2016)]

# Réinitialisez l'index
Dataframe_mortalité_2016_total = Dataframe_mortalité_2016_total.reset_index(drop=True)

# Affichez le DataFrame réinitialisé
Dataframe_mortalité_2016_total.head()
```

```python
# selection de la granularité correspondante
mortality = mortality[mortality['Granularity']== 'Total']
```

```python
Dataframe_mortalité = pd.merge(Dataframe_mortalité_2016_total, mortality, on='Country')
Dataframe_mortalité.head()
```

## Wash death en en entier

```python
Dataframe_mortalité["WASH deaths"] = Dataframe_mortalité["WASH deaths"].astype(int)
Dataframe_mortalité.head()
```

 ## Colonne taux de mortalité eau insalubre


```python
# ajout de la colonne taux de mortalité du à l'eau insalubre
Dataframe_mortalité["taux de mortalité du à l'eau insalubre (%)"] = round((Dataframe_mortalité["WASH deaths"] / Dataframe_mortalité["Population"])*100,2)
Dataframe_mortalité.head()
```

```python
Dataframe_mortalité.drop(["Year_y", "Granularity_y", "Country"], axis=1,inplace =True)
Dataframe_mortalité.head()
```
