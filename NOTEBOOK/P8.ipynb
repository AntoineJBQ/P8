## Population urbaine

```python
population.head()
```

```python
# Regrouper les donnes par pays, anne et Granularity, puis calculer la somme de la population
grouped = population.groupby(['Country', 'Year', 'Granularity'])['Population'].sum().unstack()

# # Calculer la part de la population urbaine sur la population totale
grouped['Urban Population Share (%)'] = round((grouped['Urban'] / grouped['Total']) *100 ,2)

# Rinitialiser l'index
grouped.reset_index(inplace=True)

# Fusionner les donnes calcules avec le DataFrame d'origine
population = pd.merge(population, grouped[['Country', 'Year', 'Urban Population Share (%)']], on=['Country', 'Year'], how='left')

# Afficher le DataFrame avec la nouvelle colonne
population.tail()
```

## merge avec basic and safety

```python
basic_and_safety.head()
```

```python
merged_data = population.merge(basic_and_safety, on=["Country", "Granularity", "Year"], how='left')
merged_data.head(10)
```

## merge avec political stability

```python
political_stability.head(10)
```

```python
merged_data = merged_data.merge(political_stability, on=["Country", "Year", "Granularity"], how='left')
merged_data.head(10)
```

## Selection de la granularit totale

```python
# Slectionnez les lignes o 'Granularity' est gal  'Total'
total_data_df = merged_data[merged_data['Granularity'] == 'Total']
# Affichez le DataFrame avec seulement les lignes 'Granularity' gales  'Total'
total_data_df.head()
```

## Jointures country avec le reste

```python
region_country.head(10)
```

```python
chine = region_country[region_country['COUNTRY (DISPLAY)']== 'China']
chine
```

```python
total_data_df = total_data_df.merge(region_country, left_on="Country", right_on="COUNTRY (DISPLAY)", how='left')
total_data_df.head()
```

## Cration des part de la pop ayant accs et accs safe  l'eau et non accs  l'eau

```python
total_data_df["part de la population ayant accs  l'eau"]= round(total_data_df['Population'] *( total_data_df['Population using at least basic drinking-water services (%)']/100),2)
total_data_df.head()
```

```python
total_data_df["part de la population ayant accs scuris  l'eau"]= round(total_data_df['Population'] * (total_data_df['Population using safely managed drinking-water services (%)']/100),2)
total_data_df.head()
```

```python
total_data_df["difference entre accs et accs safe"]= round((total_data_df["part de la population ayant accs  l'eau"] - total_data_df["part de la population ayant accs scuris  l'eau"]),2)
total_data_df.head()
```

```python
total_data_df["part de la population n'ayant pas un accs  l'eau"]= round(total_data_df["Population"] -(total_data_df["difference entre accs et accs safe"] + total_data_df["part de la population ayant accs scuris  l'eau"]),2)
total_data_df.head()
```

## Normalisation des donnes


```python
# x_normalized = (x - min(colonne)) / (max(colonne) - min(colonne))
# x_normalized = (x - moyenne(colonne)) / cart-type(colonne)

# Crez un MinMaxScaler pour normaliser les colonnes
scaler = MinMaxScaler()

# Normalisez les colonnes "col1" et "col2"
columns_to_normalize = ['Political_Stability', 'Urban Population Share (%)']
total_data_df[columns_to_normalize] = scaler.fit_transform(total_data_df[columns_to_normalize])

total_data_df.head()
```

# Cration du deuxime csv avec un tri sur l'anne 2016 afin de faire la jointure avec les donnes de mortalits sur cette anne

## jointure avce le fichier de mortalit

```python
# selection de l'anne 2016
Dataframe_mortalit_2016_total = total_data_df[(total_data_df['Year'] == 2016)]

# Rinitialisez l'index
Dataframe_mortalit_2016_total = Dataframe_mortalit_2016_total.reset_index(drop=True)

# Affichez le DataFrame rinitialis
Dataframe_mortalit_2016_total.head()
```

```python
# selection de la granularit correspondante
mortality = mortality[mortality['Granularity']== 'Total']
```

```python
Dataframe_mortalit = pd.merge(Dataframe_mortalit_2016_total, mortality, on='Country')
Dataframe_mortalit.head()
```

## Wash death en en entier

```python
Dataframe_mortalit["WASH deaths"] = Dataframe_mortalit["WASH deaths"].astype(int)
Dataframe_mortalit.head()
```

 ## Colonne taux de mortalit eau insalubre


```python
# ajout de la colonne taux de mortalit du  l'eau insalubre
Dataframe_mortalit["taux de mortalit du  l'eau insalubre (%)"] = round((Dataframe_mortalit["WASH deaths"] / Dataframe_mortalit["Population"])*100,2)
Dataframe_mortalit.head()
```

```python
Dataframe_mortalit.drop(["Year_y", "Granularity_y", "Country"], axis=1,inplace =True)
Dataframe_mortalit.head()
```
