## Calcul de la Population urbaine

```python
population.head()
```

```python
# Regrouper les données par pays, année et Granularity, puis calculer la somme de la population
grouped = population.groupby(['Country', 'Year', 'Granularity'])['Population'].sum().unstack()

# Calculer la part de la population urbaine sur la population totale
grouped['Urban Population Share (%)'] = round((grouped['Urban'] / grouped['Total']) *100 ,2)

# Réinitialiser l'index
grouped.reset_index(inplace=True)

# Fusionner les données calculées avec le DataFrame d'origine
population = pd.merge(population, grouped[['Country', 'Year', 'Urban Population Share (%)']], on=['Country', 'Year'], how='left')

# Afficher le DataFrame avec la nouvelle colonne
population.tail()
```

## Création des part de la pop ayant accès et accès safe à l'eau et non accès à l'eau

```python
total_data_df["part de la population ayant accès sécurisé à l'eau"]= round(total_data_df['Population'] * (total_data_df['Population using safely managed drinking-water services (%)']/100),2)
total_data_df["part de la population ayant accès à l'eau"]= round(total_data_df['Population'] *( total_data_df['Population using at least basic drinking-water services (%)']/100),2)
total_data_df["difference entre accès et accès safe"]= round((total_data_df["part de la population ayant accès à l'eau"] - total_data_df["part de la population ayant accès sécurisé à l'eau"]),2)
total_data_df["part de la population n'ayant pas un accès à l'eau"]= round(total_data_df["Population"] -(total_data_df["difference entre accès et accès safe"] + total_data_df["part de la population ayant accès sécurisé à l'eau"]),2)
total_data_df.head()
```

# Attricbution south america

```python
south_america_countries = ['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Ecuador', 'Guyana', 'Paraguay', 'Peru', 'Suriname', 'Uruguay', 'Venezuela']
```

```python
def assign_region(country, region):
    if country in south_america_countries:
        return 'South America'
    else:
        return region

total_data_df['REGION (DISPLAY)'] = total_data_df.apply(lambda row: assign_region(row['COUNTRY (DISPLAY)'], row['REGION (DISPLAY)']), axis=1)
```
